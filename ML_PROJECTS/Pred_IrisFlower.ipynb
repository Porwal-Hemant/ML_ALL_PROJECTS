```python
 import pandas as pd
 import numpy as np
 import matplotlib.pyplot as plt
 import seaborn as sns
 %matplotlib inline
```

```python
columns = ['Sepal Length' , 'Sepal width' , 'Petal Length' , 'Petal width' , 'class Labels']
df = pd.read_csv('/content/iris.data.csv' , names = columns)
df.head()
```

```python
X = df.drop('class Labels' , axis = 1 )

y = df['class Labels']
```

```python
X
```

```python
y
```

## splitting it into training set and test set

```python
from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test = train_test_split( X , y , test_size = 0.2 , random_state = 0)
```

```python
X_train
```

```python
y_train
```

## first using SUPPORT VECTOR MACHINE

```python
from sklearn.svm import SVC
svc = SVC(kernel = 'linear')
svc.fit(X_train , y_train)
```

```python
svc.predict(X_test)
```

```python
svc.score(X_test , y_test)
```

## secondly from LOGISTIC REGRESSION

```python
from sklearn.linear_model import LogisticRegression
Lr = LogisticRegression()
Lr.fit(X_train , y_train)
```

```python
Lr.predict(X_test)
```

```python
Lr.score(X_test , y_test )
```
